You are given a nested list of integers `nestedList`. Each element is either:

1. An integer.
2. A list whose elements may also be integers or other lists.

The **depth** of an integer is the number of lists it is inside of. The task is to return the sum of each integer in `nestedList` multiplied by its depth.

---

### Example 1

**Input:**  
`nestedList = [[1, 1], 2, [1, 1]]`

**Output:**  
`10`

**Explanation:**  
- Four 1's at depth 2: `1*2 + 1*2 + 1*2 + 1*2 = 8`.  
- One 2 at depth 1: `2*1 = 2`.  
- Total: `8 + 2 = 10`.

---

### Example 2

**Input:**  
`nestedList = [1, [4, [6]]]`

**Output:**  
`27`

**Explanation:**  
- One 1 at depth 1: `1*1 = 1`.  
- One 4 at depth 2: `4*2 = 8`.  
- One 6 at depth 3: `6*3 = 18`.  
- Total: `1 + 8 + 18 = 27`.

---

### Example 3

**Input:**  
`nestedList = [0]`

**Output:**  
`0`

---

## Constraints

- `1 <= nestedList.length <= 50`
- The values of the integers in the nested list are in the range `[-100, 100]`.
- The maximum depth of any integer is less than or equal to `50`.

---

## Approach

1. Use a **recursive function** to traverse the nested list:
   - If the element is an integer, multiply it by the current depth.
   - If the element is a list, recursively calculate the sum of the elements in the list at the next depth.
2. Sum up all the results to compute the final output.

This ensures efficient traversal and calculation while adhering to the depth constraints.


```python
class Solution:
    def depthSum(self, nestedList: List[NestedInteger]) -> int:
        self.total = 0
        def helper(val, d):
            for v in val:
                if v.isInteger():
                    self.total+=v.getInteger()*d
                else:
                    helper(v.getList(), d+1)
        
        helper(nestedList, 1)
        return self.total
```

```python
class Solution:
    def depthSum(self, nestedList: List[NestedInteger]) -> int:
        def helper(val, depth):
            total = 0
            for v in val:
                if v.isInteger():
                    total += v.getInteger() * depth
                else:
                    total += helper(v.getList(), depth + 1)
            return total
        
        return helper(nestedList, 1)
```
