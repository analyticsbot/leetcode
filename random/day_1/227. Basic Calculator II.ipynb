Given a string `s` representing an expression, your task is to evaluate the expression and return its value. The integer division should truncate toward zero. You are not allowed to use built-in functions like `eval()`.

### Example 1

**Input:**  
`s = "3+2*2"`

**Output:**  
`7`

---

### Example 2

**Input:**  
`s = " 3/2 "`

**Output:**  
`1`

---

### Example 3

**Input:**  
`s = " 3+5 / 2 "`

**Output:**  
`5`

---

## Constraints

- `1 <= s.length <= 3 * 10^5`
- `s` consists of integers and operators (`+`, `-`, `*`, `/`) separated by spaces.
- All integers are non-negative and within the range `[0, 231 - 1]`.
- The answer is guaranteed to fit in a 32-bit integer.


```python
class Solution:
    def calculate(self, s: str) -> int:
        curr_num = 0
        last_expr = '+'
        total = []

        # Traverse through the string
        for i in range(len(s)):
            char = s[i]
            
            # If character is a digit, build the number
            if char.isdigit():
                curr_num = curr_num * 10 + int(char)
            
            # If character is an operator or we are at the end of the string
            if char in '+-*/' or i == len(s) - 1:
                if last_expr == '+':
                    total.append(curr_num)
                elif last_expr == '-':
                    total.append(-curr_num)
                elif last_expr == '*':
                    total[-1] = total[-1] * curr_num
                elif last_expr == '/':
                    total[-1] = int(total[-1] / curr_num)
                
                # Reset for the next number
                last_expr = char
                curr_num = 0

            # Skip spaces (they are not needed for calculation)
            elif char == ' ':
                continue

        # Return the sum of the total list
        return sum(total)
```
