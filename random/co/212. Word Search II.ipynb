Given an `m x n` board of characters and a list of strings `words`, return all words on the board.  

Each word must be constructed from letters of sequentially adjacent cells, where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once in a word.  

---

#### Example 1:  
**Input:**  
`board = [["o","a","a","n"],["e","t","a","e"],["i","h","k","r"],["i","f","l","v"]]`,  
`words = ["oath","pea","eat","rain"]`  
**Output:**  
`["eat","oath"]`  

---

#### Example 2:  
**Input:**  
`board = [["a","b"],["c","d"]]`,  
`words = ["abcb"]`  
**Output:**  
`[]`  

---

#### Constraints:  
- \( m == \text{board.length} \)  
- \( n == \text{board}[i].\text{length} \)  
- \( 1 \leq m, n \leq 12 \)  
- \( \text{board}[i][j] \) is a lowercase English letter.  
- \( 1 \leq \text{words.length} \leq 3 \times 10^4 \)  
- \( 1 \leq \text{words}[i].\text{length} \leq 10 \)  
- \( \text{words}[i] \) consists of lowercase English letters.  
- All the strings of `words` are unique.  


```python
class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        # Helper function to build a Trie
        def build_trie():
            trie = {}
            for word in words:
                node = trie
                for char in word:
                    if char not in node:
                        node[char] = {}
                    node = node[char]
                node['#'] = True  # Mark the end of a word
            return trie

        # DFS function to search for words
        def dfs(x, y, node, path):
            char = board[x][y]
            if char not in node:
                return
            next_node = node[char]
            if '#' in next_node:
                found_words.add(path + char)

            # Mark the cell as visited
            board[x][y] = '*'
            for dx, dy in directions:
                nx, ny = x + dx, y + dy
                if 0 <= nx < m and 0 <= ny < n and board[nx][ny] != '*':
                    dfs(nx, ny, next_node, path + char)
            # Restore the cell
            board[x][y] = char

        # Build the Trie
        trie = build_trie()
        m, n = len(board), len(board[0])
        directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        found_words = set()

        # Start DFS from each cell
        for i in range(m):
            for j in range(n):
                dfs(i, j, trie, "")

        return list(found_words)
```
