Given the root of a binary tree, return the length of the diameter of the tree.

The **diameter** of a binary tree is the length of the longest path between any two nodes in a tree. This path may or may not pass through the root.

The length of a path between two nodes is represented by the number of edges between them.

---

### Example 1:

**Input:**  
`root = [1,2,3,4,5]`  
**Output:**  
`3`  
**Explanation:**  
3 is the length of the path `[4,2,1,3]` or `[5,2,1,3]`.

---

### Example 2:

**Input:**  
`root = [1,2]`  
**Output:**  
`1`  

---

### Constraints:
- The number of nodes in the tree is in the range `[1, 10^4]`.
- `-100 <= Node.val <= 100`


```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        self.diameter = 0
        
        def height(node):
            if not node:
                return 0  # Base case: height of an empty subtree is 0
            
            # Recursively calculate the height of left and right subtrees
            left_height = height(node.left)
            right_height = height(node.right)
            
            # Update the diameter: maximum path through the current node
            self.diameter = max(self.diameter, left_height + right_height)
            
            # Return the height of the subtree rooted at this node
            return 1 + max(left_height, right_height)
        
        height(root)  # Start DFS traversal from the root
        return self.diameter
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode], target: Optional[TreeNode]) -> int:
        self.diameter = 0
        self.target_diameter = 0  # To track the diameter through the specific target node
        
        def height(node):
            if not node:
                return 0  # Base case: height of an empty subtree is 0
            
            # Recursively calculate the height of left and right subtrees
            left_height = height(node.left)
            right_height = height(node.right)
            
            # Update the global diameter
            self.diameter = max(self.diameter, left_height + right_height)
            
            # If the current node is the target, update the target's diameter
            if node == target:
                self.target_diameter = left_height + right_height
            
            # Return the height of the subtree rooted at this node
            return 1 + max(left_height, right_height)
        
        height(root)  # Start DFS traversal from the root
        return self.target_diameter
```
