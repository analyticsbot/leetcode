Given an m x n matrix, return true if the matrix is Toeplitz. Otherwise, return false. A matrix is Toeplitz if every diagonal from top-left to bottom-right has the same elements.

### Example 1:
**Input**: `matrix = [[1,2,3,4],[5,1,2,3],[9,5,1,2]]`  
**Output**: `true`  
**Explanation**: In the above grid, the diagonals are: "[9]", "[5, 5]", "[1, 1, 1]", "[2, 2, 2]", "[3, 3]", "[4]". In each diagonal all elements are the same, so the answer is True.

### Example 2:
**Input**: `matrix = [[1,2],[2,2]]`  
**Output**: `false`  
**Explanation**: The diagonal "[1, 2]" has different elements.

### Constraints:
- `m == matrix.length`
- `n == matrix[i].length`
- `1 <= m, n <= 20`
- `0 <= matrix[i][j] <= 99

### Follow-up:
- What if the matrix is stored on disk, and the memory is limited such that you can only load at most one row of the matrix into the memory at once?
- What if the matrix is so large that you can only load up a partial row into the memory at once?


```python
class Solution:
    def isToeplitzMatrix(self, matrix: List[List[int]]) -> bool:
        # Start from the second row and second column
        m,n = len(matrix), len(matrix[0])

        for i in range(m):
            for j in range(n):
                if i+1<m and j+1<n:
                    if matrix[i][j]!=matrix[i+1][j+1]:
                        return False

        return True
```

```python
class Solution:
    def isToeplitzMatrix(self, matrix: List[List[int]]) -> bool:
        m, n = len(matrix), len(matrix[0])

        # Check diagonals starting from first row
        for j in range(n):
            value = matrix[0][j]
            i, k = 0, j
            while i + 1 < m and k + 1 < n:
                if matrix[i + 1][k + 1] != value:
                    return False
                i += 1
                k += 1
        
        # Check diagonals starting from first column
        for i in range(1, m):
            value = matrix[i][0]
            j, k = i, 0
            while j + 1 < m and k + 1 < n:
                if matrix[j + 1][k + 1] != value:
                    return False
                j += 1
                k += 1

        return True
```

```python
## What if the matrix is stored on disk, and the memory is limited such that you can only load at most one row of the matrix into the memory at once?

class Solution:
    def isToeplitzMatrix(self, matrix: List[List[int]]) -> bool:
        m, n = len(matrix), len(matrix[0])

        # We will keep track of the previous row while reading each row
        previous_row = None

        # Iterate through each row of the matrix
        for i in range(m):
            # Read the current row
            current_row = matrix[i]

            # Compare the current row with the previous row
            if previous_row is not None:
                # Check each element with the element diagonally down-right
                for j in range(n - 1):
                    if previous_row[j] != current_row[j + 1]:
                        return False
            
            # Set the current row as the previous row for the next iteration
            previous_row = current_row

        return True
```

```python
## What if the matrix is so large that you can only load up a partial row into the memory at once?
class Solution:
    def isToeplitzMatrix(self, matrix: List[List[int]], chunk_size: int) -> bool:
        m, n = len(matrix), len(matrix[0])

        # Initialize a buffer for the previous row (partial or full depending on chunk size)
        previous_row_chunk = None

        # Process matrix row by row
        for i in range(m):
            current_row_chunk = matrix[i]

            # Compare current row with previous row
            if previous_row_chunk is not None:
                # We can compare the diagonal elements from previous and current chunk
                for j in range(min(len(current_row_chunk), len(previous_row_chunk) - 1)):
                    if previous_row_chunk[j] != current_row_chunk[j + 1]:
                        return False

            # Save the current chunk to be used in the next iteration
            previous_row_chunk = current_row_chunk

        return True
```
